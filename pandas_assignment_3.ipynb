```python
# which commodity was having the highest opening price for the day

data.loc[data["open"].idxmax(), "commodity"]
```

```python
# which commodity price reached max during the trading session.

max_price = data.groupby("commodity")
max_price.max().sort_values("open",ascending = False)
```

```python
# what is the average  markers opening price for the Oat in 2010

data["year"] = pd.DatetimeIndex(data["date"]).year
data
```

```python
ac = data.loc[data["commodity"] == "Oat",["commodity","open","year"]]
con = ac[ac["year"] == 2010]
result = con["open"].mean()
print("the average  markers opening price for the Oat in 2010 : ",result)
```

```python
# what is the max and min number of contracts traded during the 2009

year = data.loc[data["year"] == 2009,["volume","year"]]
max_contracts = year["volume"].max()
min_contracts = year["volume"].min()
print("max number of contracts traded during the 2009: ",max_contracts)
print()
print("min number of contracts traded during the 2009: ",min_contracts)
```

```python
# which is the max and min number of contracts traded

max_contracts = data.loc[data["volume"].idxmax(),"year"]
min_contracts = data.loc[data["volume"].idxmin(),"year"]
print("max number of contracts traded : ",max_contracts)
print()
print("min number of contracts traded : ",min_contracts)
```

```python
# which commodity have the average lowest opening price around 2001 to 2005

data[(data["year"] >= 2001) & (data["year"] <= 2005)]
```

```python
# which commodity have the best average markets closing price

data.pivot_table(index = "commodity",values = "close",aggfunc = "mean")
```

```python
# which year corn hav the highest and lowest number of contracts traded

corn = data[data["commodity"] == "Corn"]
max_corn = corn.loc[corn["volume"].idxmax(),"year"]
min_corn = corn.loc[corn["volume"].idxmin(),"year"]
print("max year corn: ",max_corn)
print()
print("min year corn: ",min_corn)
```
