```python
# how many unique brands are there in the given data

data["brand"].nunique()
```

```python
# what is the average ram size of all the collected data

data["ram(GB)"].mean()
```

```python
# which phone is having the max and min battery capacity

max_battery = data.loc[data["battery"].idxmax(),"phone_name"]
print("max battery: ",max_battery)

min_battery = data.loc[data["battery"].idxmin(),"phone_name"]
print("min battery: ",min_battery)
```

```python
# what is the maximum battery capacity provided by huawei company

phone = data[data["brand"] == "Huawei"]
max_battery = phone["battery"].max()
print("maximum battery capacity provided by huawei company: ",max_battery)
```

```python
# which phone brand is providing the highest and lowest ram

phone = data.loc[data["ram(GB)"].idxmax(),"brand"]
print("highest phone brand ram: ",phone)

phone = data.loc[data["ram(GB)"].idxmin(),"brand"]
print("lowest phone brand ram: ",phone)
```

```python
# how many phones has android 7.0 as there operating system

data[data["os"] == "Android 7.0"]
```

```python
# which phone is the oldest

data.loc[data["announcement_date"].idxmin(),"phone_name"]
```

```python
# which brand produced max number of phone in the given data

brand = data["brand"].value_counts()
max_phone = brand.idxmax()
print(max_phone)
```

```python
# how many phones supports Li-po and Li-ion Battery 

battery_phone = data[(data["battery_type"] == "Li-Po") | (data["battery_type"] == "Li-Ion")]
print("the no of phones supports Li-po and Li-ion Battery : ",len(battery_phone))
```

```python
# how many sony phone supports 720x1280 resolution

phone = data[data["brand"] == "Sony"]
resolution = phone[phone["resolution"] == "720x1280"]
print("There are",len(resolution),"sony phone supports 720x1280 resolution : ")
```
